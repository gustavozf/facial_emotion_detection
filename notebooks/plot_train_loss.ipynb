```python
import os

import pandas as pd
import matplotlib.pyplot as plt
```

```python
BASE_PATH = '/media/zanoni/Extreme SSD/facial_emotion_recognition/outputs'
NO_DATA_AUG_LOGS_PATH = f'{BASE_PATH}/v004_v2/logs.csv'
WEAK_DATA_AUG_LOGS_PATH = f'{BASE_PATH}/v007/logs.csv'
STRONG_DATA_AUG_LOGS_PATH = f'{BASE_PATH}/v008/logs.csv'
STRONG_DATA_AUG_LRS_LOGS_PATH = f'{BASE_PATH}/v010/logs.csv'
TARGET_COLS = ['loss', 'val_loss']
```

```python
no_data_aug_df = pd.read_csv(NO_DATA_AUG_LOGS_PATH)
no_data_aug_df[TARGET_COLS].plot()
plt.title('EfficientNetv2B1: No Data Aug.')
plt.xlabel('Epochs')
plt.ylabel('Loss')
```

```python
weak_data_aug_df = pd.read_csv(WEAK_DATA_AUG_LOGS_PATH)
weak_data_aug_df[TARGET_COLS].plot()
plt.title('EfficientNetv2B1: Weak Data Aug.')
plt.xlabel('Epochs')
plt.ylabel('Loss')
```

```python
strong_data_aug_df = pd.read_csv(STRONG_DATA_AUG_LOGS_PATH)
strong_data_aug_df[TARGET_COLS].plot()
plt.title('EfficientNetv2B1: Strong Data Aug.')
plt.xlabel('Epochs')
plt.ylabel('Loss')
```

```python
strong_data_aug_lrs_df = pd.read_csv(STRONG_DATA_AUG_LRS_LOGS_PATH)
strong_data_aug_lrs_df[TARGET_COLS].plot()
plt.title('EfficientNetv2B1: Strong Data Aug. + LR Scheduling')
plt.xlabel('Epochs')
plt.ylabel('Loss')
```
